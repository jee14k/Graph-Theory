### NetworkX Package: 
#### It is a python package for the creation, manipulation and the study of the structure, dynamics and functions of complex networks. It provides the following:
* Tools for the study of the structure and dynamics of social, biological and infrastructure networks
* A standard programming interface and graph implementation that is suitable for many applications
* A rapid development environment for collaboratve, multidisciplinary projects
* An interface to existing numerical algorithms and code written in C, C++ and FORTRAN
* The ability to painlessly work with large nonstandard data sets

#### With NetwokX, you can load and store networks in standard and nonstandard data formats, generate many types of random and classic networks, analyze network structure, build network models, design new network algorithms, draw networks and so on.

### Directed Graph:

#### A graph G in which every edge is a directed edge (edge with a direction) is known as directed graph or digraph. A directed graph obtained from an ordinary graph G by assigning directions to its edges is called an orientation of G. Here, G is called the underlying graph of the directed graph. 


````python
GD= nx.DiGraph()
GD.add_nodes_from(['1','2','3','4','5'])
GD.add_edge('1','2')
GD.add_edge('2','3')
GD.add_edge('3','4')
GD.add_edge('4','1')
GD.add_edge('4','5')
nx.draw(GD, with_labels=True, node_color="yellow" ,node_size=400)
plt.show()
````

### Drawing graphs using coordinates:

````python
HG= nx.house_graph() #explicitly set positions
pos={0:(0,0),
    1:(1,0),
    2:(0,1),
    3:(1,1),
    4:(0.5, 2.0)}
nx.draw_networkx_nodes(HG, pos, node_size=300, nodelist=[4], node_color='r')
nx.draw_networkx_nodes(HG, pos, node_size=300, nodelist=[0,3], node_color='b')
nx.draw_networkx_nodes(HG, pos, node_size=300, nodelist=[1,2], node_color='g')
nx.draw_networkx_edges(HG, pos, alpha=0.75, width=3)
plt.axis('on')
plt.show()
````
